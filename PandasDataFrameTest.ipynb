```python
# Series的算数运算和对其操作
ser7 = pd.Series(np.arange(10))
ser8 = pd.Series(np.arange(5))
ser7+ser8
```

```python
# DataFrame的算数运算和对其操作
ndarray = np.random.randn(2,2)
frame7 = pd.DataFrame(ndarray)
ndarray = np.random.randn(3,3)
frame8 = pd.DataFrame(ndarray)
frame7+frame8
frame7.add(frame8,fill_value=10)
```
