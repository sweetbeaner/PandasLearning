```python
import numpy as np,pandas as pd
```

```python
io = r'C:\Users\chen\Desktop\蓄水量-6.3.3节\2016年\【日损耗量】2016年蓄水量变化量汇总.xlsx'
io
A = '10cm'
B = '20cm'
C = '30cm'
D = '40cm'
E = '50cm'
F = '60cm'
G = '70cm'
H = '80cm'
I = '90cm'
J = '100cm'
K = '120cm'
L = '140cm'
M = '160cm'
N = '180cm'
O = '200cm'
writer = pd.ExcelWriter(r'C:\Users\chen\Desktop\蓄水量-6.3.3节\2016年\2016年蓄水量求和计算.xlsx', engine='xlsxwriter')
```

```python
s_name = '刺槐1'
data = pd.read_excel(io, sheet_name = s_name,index_col = '日期')
print(data)
list = [A,B,C,D,E,F,G,H,I,J,K,L,M,N,O]
col = ['sum+','sum-']
results = pd.DataFrame(index=list,columns = col)
# print(results)
for i in list:
    ser1 = data[data[i]>=0][i]
    max = ser1.sum()
    ser2 = data[data[i]<0][i]
    min = ser2.sum()
    results.loc[i,'sum+']=max
    results.loc[i,'sum-']=min
print(results)
results.to_excel(writer, sheet_name=s_name, index=True)
```
