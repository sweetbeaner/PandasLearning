```python
import numpy as np, pandas as pd
```

```python
# 创建一个测试DataFrame
frame = pd.DataFrame(np.random.randn(10,10),columns=['a','b','c','d','e','f','g','h','j','k'])
frame
```

```python
# 单列求和
frame['c'].sum()
```

```python
# 单行求和
frame.sum(axis = 'columns')
```

```python
# NaN值处理，顾及NaN值
frame
frame.sum(axis=1,skipna=False)
```

```python
frame
```

```python
# 返回每一列最大值的index
frame.idxmax()
```

```python
# 第n个元素返回所在列前n个元素的累加值
frame.cumsum()
```

```python
# 非归约也非累计，产生汇总统计结果
descFrame = frame.describe()
print(descFrame)
```

```python
# 非NaN的值的个数
frame.count()
```

```python
# 索引位置和标签
print(descFrame.idxmax())# 返回index标签
descFrame['b'].argmax()# 返回索引整数
```
